```python
from sklearn.datasets import make_moons
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X,y = make_moons(n_samples=30000,noise=0.15,random_state=2)

X_train,X_teste,y_train,y_teste = train_test_split(X,y,random_state=2,test_size=0.3,shuffle=True)

# Criar classificadores

log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clt = SVC()

voting_clf = VotingClassifier(
    estimators=[('lr',log_clf),('rf',rnd_clf),('svm',svm_clt)],
    voting='hard'
)

voting_clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf,rnd_clf,svm_clt,voting_clf):
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_teste)
    print(clf.__class__.__name__,accuracy_score(y_teste,y_pred))
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    max_samples=100, bootstrap=True, n_jobs=-1
)

bag_clf.fit(X_train,y_train)
```

```python
y_pred = bag_clf.predict(X_teste)
print(accuracy_score(y_teste,y_pred))

tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train,y_train)

y_pred_tree = tree_clf.predict(X_teste)
print(accuracy_score(y_teste,y_pred_tree))
```

```python
# ADA BOOST

from sklearn.ensemble import AdaBoostClassifier
ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1),n_estimators=200,
    algorithm="SAMME", learning_rate=0.5
)

ada_clf.fit(X_train,y_train)
```

```python
from sklearn.datasets import fetch_openml
from sklearn.linear_model import SGDClassifier
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# 1. Carregar MNIST
print("Iniciando programa")
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
print("MNIST carregado!")
```

```python
X, y = mnist["data"], mnist["target"].astype(np.uint8)
X_train, X_test, y_train, y_test = X[:1000], X[1000:], y[:1000], y[1000:]


# Criar classificadores
log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clt = SVC()

voting_clf = VotingClassifier(
    estimators=[('lr',log_clf),('rf',rnd_clf),('svm',svm_clt)],
    voting='hard'
)

voting_clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf,rnd_clf,svm_clt,voting_clf):
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__,accuracy_score(y_test,y_pred))
```
